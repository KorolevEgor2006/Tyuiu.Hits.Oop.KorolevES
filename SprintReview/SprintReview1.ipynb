```polyglot-notebook

public class Product
{
    private string name;
    private decimal price;
    private int quantity;
    protected string category;
    internal string description;
    private static decimal TotalPrice;

    public string Name
    {
        get{return name;}
        set{name=value;}
    }
    public string Category
    {
        get{return category;}
        set{category=value;}
    }
    public decimal Price
    {
        get{return price;}
        private set{
            if(value>=0)price=value;
            else throw new ArgumentOutOfRangeException("Цена не может быть отрицательной.");
        }
    }
    public int Quantity
    {
        get{return quantity;}
        private set{
            if(value>=0)quantity=value;
            else throw new ArgumentOutOfRangeException("Количество не может быть отрицательной.");
        }
    }

    public Product(string _name, string _category, decimal _price)
    {
        Name=_name;
        Category=_category;
        Price=_price;
        TotalPrice+=price;
    }
    public static decimal GetTotalPrice()
    {
        return TotalPrice;
    }

    public void UpdateQuantity(int amount)
    {
        int last =Quantity;
        Quantity=amount;
        Console.WriteLine($"Количество товара {Name} изменилось с {last} на {Quantity}.");
    }

    public string GetProductInfo()
    {
        return $"Товар {Name}: {description}. Категория: {category}. Количество: {Quantity}. Цена:{Price}";
    }
}
public class Seller
{
    private string name;
    private string employeeId;
    protected decimal salary;
    internal string contactInfo;
    public List<Product> SellerProducts;
    public string Name
    {
        get{return name;}
        set{name=value;}
    }
    public string EmployeeId
    {
        get{return employeeId;}
        set{employeeId=value;}
    }
    public decimal Salary
    {
        get{return salary;}
        set{
            if(value>=0)salary=value;
            else throw new ArgumentOutOfRangeException("Стоимость проданного не может быть отрицательной.");
        }
    }
    public Seller(string _name,string _employeeId)
    {
        Name=_name;
        EmployeeId=_employeeId;
        Salary=0;
        SellerProducts= new List<Product>();
    }
    public void AddProduct(Product product)
    {
        SellerProducts.Add(product);
        Console.WriteLine($"В список продавца добавлен товар {product.Name}");
    }
    public void SellProduct(Product product, int quantity)
    {
        product.UpdateQuantity(product.Quantity-quantity);
        salary+=quantity*product.Price;
    }
    public string GetSellerInfo()
    {
        return $"Продавец {Name} ({employeeId}).Продал на сумму: {salary}. Контактная информация:{employeeId}";
    }
}
public class Store
{
    private string storeName;
    private string location;
    public string storeHouse;
    private List<Seller> StoreSellers;
    public string StoreName
    {
        get{return storeName;}
        set{storeName=value;}
    }
    public string Location
    {
        get{return location;}
        set{location=value;}
    }
    public Store(string _storeName,string _location)
    {
        StoreName=_storeName;
        Location=_location;
        StoreSellers= new List<Seller>();
    }
    public void AddSeller(Seller seller)
    {
        StoreSellers.Add(seller);
    }

    public void ListProducts()
    {
        Console.WriteLine($"Товары в магазине {storeName}:");
        foreach(Seller seller in StoreSellers)
        {
            foreach(Product product in seller.SellerProducts)
            {
                Console.WriteLine($"{product.Name}: категория: {product.Category},цена: {product.Price} , количество: {product.Quantity}");
            }
        }
    }
    public string GetStoreInfo()
    {
        return $"Магазин {storeName}: Расположение: {location}, Часы работы: {storeHouse}";
    }
}
Product bread= new Product("хлеб","eда",100);
bread.description="булочное изделие";
bread.UpdateQuantity(5);
Product milk= new Product("молоко","eда",50);
milk.description="кисломолочный-продукт";
milk.UpdateQuantity(3);
Console.WriteLine(milk.GetProductInfo());
Seller seller0= new Seller("Иванов Иван Иванович","1");
seller0.contactInfo="телефон: +7 000 000 00 00";
seller0.AddProduct(bread);
seller0.AddProduct(milk);
Console.WriteLine(seller0.GetSellerInfo());
Store store0 = new Store("Магнит","Ул Пермякова 8");
store0.storeHouse="7:00 -19:00";
store0.AddSeller(seller0);
store0.ListProducts();
Console.WriteLine(store0.GetStoreInfo());
Console.WriteLine($"Общая стоимость всех единиц товаров:{Product.GetTotalPrice()}")
```
